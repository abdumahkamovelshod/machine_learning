```python
from fastai.vision.all import *
```

```python
path=Path('/kaggle/input/pnevmoniya/train')
path.ls()
```

```python
fls=get_image_files(path)
fls
```

```python
failed=verify_images(path)
failed
```

```python
pt=DataBlock(
blocks=(ImageBlock,CategoryBlock),
get_items=get_image_files,
splitter=RandomSplitter(valid_pct=0.2,seed=42),
get_y=parent_label,
item_tfms=Resize(224)
)
```

```python
dls=pt.dataloaders(path)
```

```python
dls.train.show_batch(max_n=16,nrows=4)
```

```python
learn=cnn_learner(dls,resnet34,metrics=accuracy)
learn.fine_tune(4)
```

```python
interp=ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(3,nrows=1)
```

```python
sub=pd.read_csv('/kaggle/input/pnevmoniya/sample_solution.csv')
sub.head()
```

```python
result = []
for i in range(len(sub)):
    label = sub.loc[i, 'id']
    img = PILImage.create(Path(f"/kaggle/input/pnevmoniya/test/{label}"))
    pred, pred_id, probs = learn.predict(img)
    pred = 1 if pred=='PNEUMONIA' else 0 
    result.append(pred)

    

solution = pd.read_csv('/kaggle/input/pnevmoniya/sample_solution.csv')
solution['labels'] = result
solution.head()
```
